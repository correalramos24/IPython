# Pyplot from Matplotlib

Basic plotting package for Python.

Pyplot uses `Figures` as the main object that contains one or more `Axes`, which are the regions where the plot is printed itself.



```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

## Basic example


```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the axes.
```
