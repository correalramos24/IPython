# PANDAS

Pandas is a library to manage dataset for data analysis.

The basic unit is the **DataFrame**, which is a table (2D) composed of:
* **series** (columns, vertical axis) with a column lable
* **records**(rows, horitzontal axis) with a unique Id named index


```python
import pandas as pd
```

## Reading and exporting data
With pandas you can read data from files types, with read_{type} methods, and write with to_{type} methods.

![](https://pandas.pydata.org/docs/_images/02_io_readwrite.svg)

```python
# You can use to convert one file to other:
people_ds = pd.read_csv("people.csv")
people_ds.to_html("people.html")
```

## View data
We can access data from the whole dataset, by series and by records.

Also there are "main views" for a quick-view of the dataFrame:

```python
# Display a main view, the first and the lasts elements:
#people_ds
people_ds.dtypes
people_ds.info()
people_ds.head(10)
#people_ds.tail(10)
```

## Select data
About the records and series, we can use the operator `[]` for select some
series. We can also filter with a sweet semantic:

```python
# Select series:
names = people_ds["First Name"]
names_and_sex = people_ds[["First Name", "Sex"]]
names_and_sex.head(3)
print("Types of the slices", type(names), type(names_and_sex))
print("Shape views", names.shape, names_and_sex.shape)

# Access & filter: We can use == != < <= and more logical operators
males = people_ds[people_ds["Sex"] == "Male"]
males

# Select records:
# 1) Pick from the Sex == Male just the Firt Name
first_male_index = people_ds.loc[people_ds["Sex"] == "Male", "First Name"]
first_male_index.head(3)

# 2) Pick with the rows and columns Id
first_elements = people_ds.iloc[0:8, 2:5]
first_elements

# 3) Use python slices over a serie
people_ds["First Name"][0:4]
```
